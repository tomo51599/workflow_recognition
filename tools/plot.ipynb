```python
import numpy as np

def calculate_metrics(confusion_matrix):
    # 真の正、偽の正、偽の負を計算
    TP = np.diag(confusion_matrix)  # 対角成分が真の正
    FP = np.sum(confusion_matrix, axis=0) - TP
    FN = np.sum(confusion_matrix, axis=1) - TP

    # Precision, Recall, F1-Scoreを計算
    precision = TP / (TP + FP)
    recall = TP / (TP + FN)
    f1_score = 2 * (precision * recall) / (precision + recall)

    # NaNを0で置換（発生する可能性がある0除算を防ぐ）
    precision = np.nan_to_num(precision)
    recall = np.nan_to_num(recall)
    f1_score = np.nan_to_num(f1_score)
    
    return precision, recall, f1_score

# 混合行列の読み込み
matrix_1 = np.array([
    [342, 1, 6, 0, 0, 54],
    [8, 145, 91, 0, 0, 30],
    [0, 0, 732, 5, 0, 0],
    [1, 1, 77, 779, 88, 0],
    [1, 5, 39, 262, 746, 13],
    [14, 126, 380, 0, 0, 1734]
])

matrix_2 = np.array([
    [329, 53, 21, 0, 0, 0],
    [0, 128, 146, 0, 0, 0],
    [0, 0, 679, 58, 0, 0],
    [0, 0, 28, 918, 0, 0],
    [0, 0, 0, 144, 890, 32],
    [0, 0, 0, 0, 0, 2254]
])

# 計算実行
precision_1, recall_1, f1_score_1 = calculate_metrics(matrix_1)
precision_2, recall_2, f1_score_2 = calculate_metrics(matrix_2)

print("Metrics for Matrix 1:")
print("Precision:", precision_1)
print("Recall:", recall_1)
print("F1 Score:", f1_score_1)

print("\nMetrics for Matrix 2:")
print("Precision:", precision_2)
print("Recall:", recall_2)
print("F1 Score:", f1_score_2)

```
